# Dimensionality Reduction

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
np.random.seed(50)
x1 = 2 * np.random.randn(250)
x2 = 0.8 * x1 + 1 + 0.5 * np.random.randn(250)
X = np.concatenate((x1.reshape(-1, 1), x2.reshape(-1, 1)), axis=1)

plt.scatter(x1, x2, c='b', alpha=0.5)
plt.title('X')
plt.show()
```

## Principal Component Analysis (PCA)

### Singular Value Decomposition (SVD): $X = U \Sigma V^{T}$
$V = \begin{pmatrix}
| & | && | \\
c_{1} & c_{2} & ... & c_{n} \\
| & | && | \\
\end{pmatrix}$
### projection: $X_{project} = XW_{d}$
where $W_{d}$ is the first $d$ columns of $V$

```python
X_centered = X - X.mean(axis=0)
U, s, Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
```

```python
x = np.linspace(-8.5, 7.5, 100)
plt.scatter(X_centered[:, 0], X_centered[:, 1], c='b', alpha=0.5)
plt.plot(x, (c1[1] / c1[0]) * x, label='c1', c='k')
plt.plot(x, (c2[1] / c2[0]) * x, label='c2', c='k', ls='--')
plt.legend(loc='upper left')
plt.title('X Centered')
plt.xlim(-8.2, 7.5)
plt.ylim(-6.8, 7.5)
plt.show()
```

```python
W_d = Vt.T[:, :1]
X_project = X_centered.dot(W_d)
fig, ax = plt.subplots(1)
fig.set_figheight(1.5)
plt.scatter(X_project, np.zeros((250, 1)), c='b', alpha=0.5)
ax.set_yticks([])
plt.show()
```

### Scikit-Learn

```python
from sklearn.decomposition import PCA
```

```python
# PCA centers the X itself
pca = PCA(n_components=1)
X_project = pca.fit_transform(X)
```

```python
fig, ax = plt.subplots(1)
fig.set_figheight(1.5)
plt.scatter(X_project, np.zeros((250, 1)), c='b', alpha=0.5)
ax.set_yticks([])
plt.show()
```

### Explained Variance Ratio

```python
print(pca.explained_variance_ratio_)
```

### Right Number of Dimensions
it is preferred to choose the number of dimensions that add up to a large portion of the variance; like 95%

#### 1. cumulative sum:

```python
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print('preferred dimension:', d)
```

then you should fit the `PCA()` with `n_components = d` and transform it

#### 2. setting a float between $0.0$ and $1.0$ as `n_components`

```python
pca = PCA(n_components=0.95)
X_project = pca.fit_transform(X)
```

#### 3. plotting *explained variance ratio* vs *number of dimensions*
it is recommended when the number of dimensions of data is really high

```python
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784')

X = mnist.data
y = mnist.target
```

```python
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.plot(cumsum, c='b')
plt.grid()
plt.show()
```

### Compression

### recovery: $X_{recovered} = X_{project} W_{d}^{T}$

---
alternatively, you can use `inverse_transform` to recover with *Scikit-Learn*

```python
pca = PCA()
pca.fit(X)
cumsum = np.cumsum(pca.explained_variance_ratio_)
d = np.argmax(cumsum >= 0.95) + 1
print('preferred dimension:', d)
```

```python
some_digits_before_compression = np.array(X.iloc[0:4]).reshape(4 * 28, 28)

pca = PCA(n_components=154)
X_project = pca.fit_transform(X)
X_recovered = pca.inverse_transform(X_project)

some_digits_after_compression = X_recovered[0:4].reshape(4 * 28, 28)
```

```python
plt.subplots(1, 2)
plt.subplot(1, 2, 1)
plt.imshow(some_digits_before_compression, cmap=plt.cm.binary)
plt.title('before compression')
plt.subplot(1, 2, 2)
plt.imshow(some_digits_after_compression, cmap=plt.cm.binary)
plt.title('after compression')
plt.show()
```

### Randomized PCA
it finds an aproximation of the first $d$ principal components

---
in order to achieve that, we should set `svd_solver='randomized'` in `PCA()`

```python
rnd_pca = PCA(n_components=154, svd_solver='randomized')
X_project = rnd_pca.fit_transform(X)
```

### Incremental PCA (IPCA)
it doesn't use all the instances to fit the PCA, but a mini-batch at each step

```python
from sklearn.decomposition import IncrementalPCA
```

```python
n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X, n_batches):
    inc_pca.partial_fit(X_batch)
X_project = inc_pca.transform(X)
```

## Kernel PCA (kPCA)

```python
from mpl_toolkits import mplot3d
from sklearn.datasets import make_swiss_roll

X, y = make_swiss_roll(n_samples=800)

fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
a = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.cool_r)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, -70)
plt.colorbar(a, shrink=0.6)
plt.show()
```

```python
from sklearn.decomposition import KernelPCA
```

```python
# linear kernel (default); this equals using PCA()
pca = KernelPCA(n_components=2, kernel='linear')
X_project = pca.fit_transform(X)

plt.scatter(X_project[:, 0], X_project[:, 1], c=y, cmap=plt.cm.cool_r)
plt.colorbar()
plt.show()
```

```python
# kernel PCA (rbf kernel)
rbf_pca = KernelPCA(n_components=2, kernel='rbf', gamma=0.05)
X_project_rbf = rbf_pca.fit_transform(X)

plt.scatter(X_project_rbf[:, 0], X_project_rbf[:, 1], c=y, cmap=plt.cm.cool_r)
plt.colorbar()
plt.show()
```

```python
# kernel PCA (polynomial kernel)
poly_pca = KernelPCA(n_components=2, kernel='poly', coef0=1, degree=3)
X_project_poly = poly_pca.fit_transform(X)

plt.scatter(X_project_poly[:, 0], X_project_poly[:, 1], c=y, cmap=plt.cm.cool_r)
plt.colorbar()
plt.show()
```

```python
# kernel PCA (sigmoid kernel)
sig_pca = KernelPCA(n_components=2, kernel='sigmoid', gamma=8e-4)
X_project_sig = sig_pca.fit_transform(X)

plt.scatter(X_project_sig[:, 0], X_project_sig[:, 1], c=y, cmap=plt.cm.cool_r)
plt.colorbar()
plt.show()
```

### Best Kernel and Hyperparameters

#### 1. fitting the best model
dimensionality reduction is kind of an *unsupervised learning algorithm* (we can not measure how good we reduced dimensions), but we can search for the *kernel* and *hyperparameters* that helps us **fit the best classification or regression task** we have (which is a *supervised learning algorithm*)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
```

```python
svm_reg = Pipeline([
    ('kpca', KernelPCA(n_components=2)),
    ('svr', SVR(kernel='rbf'))
])

param_grid = [{
    'kpca__gamma': [0.01, 0.03, 0.05, 0.07, 0.1],
    'kpca__kernel': ['rbf', 'sigmoid']
}]

grid_search = GridSearchCV(estimator=svm_reg, param_grid=param_grid, cv=3, scoring='neg_mean_squared_error')
grid_search.fit(X, y)

print(grid_search.best_params_)
```

#### 2. minimizing reconstruction error
*reconstruction error* is the mean squared distance between the original data and the recovered (reconstructed) data

when we use a kernel trick, the reconstructed data is in infinite-dimensional space and, and we should use the *pre-image*, instead

```python
from sklearn.metrics import mean_squared_error
```

```python
rbf_pca = KernelPCA(n_components=2, kernel='rbf', gamma=0.05, fit_inverse_transform=True)
X_project = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_project)

print('reconstruction error:', mean_squared_error(X, X_preimage))
```

## Locally Linear Embedding (LLE)

LLE tries to model each instance as a linear function of its *$k$ nearest neigbors*, and then reconstructing each instance with lower dimensions

```python
from sklearn.manifold import LocallyLinearEmbedding
```

```python
lle = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_reduced = lle.fit_transform(X)

plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap=plt.cm.cool_r)
plt.colorbar()
plt.show()
```
