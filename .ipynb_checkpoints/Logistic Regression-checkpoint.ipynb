# Logistic Regression

## loading Iris dataset

```python
import numpy as np, matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(iris.keys())
```

```python
print(iris.feature_names)
print(iris.target_names)
```

```python
X = np.copy(iris['data'][:, 3:])
y = (iris['target'] == 2).astype(int)   # virginica predictor
```

## Scikit-Learn
*virginica* predictor - based on *petal length* and *petal width*

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
log_reg.fit(X, y)

color, marker = [], []
for i in range(len(y)):
    if (y[i] == 1):
        color.append('g')
        marker.append('o')
    else:
        color.append('b')
        marker.append('s')
        
x = np.linspace(0, 3, 100)
y_proba = log_reg.predict_proba(x.reshape(-1, 1))
plt.plot(x, y_proba[:, 1], label='Virginica', c='g')
plt.plot(x, y_proba[:, 0], label='Not Virginica', c='b', ls='--')
for x_s, y_s, c, m in zip(X, y, color, marker):
    plt.scatter(x_s, y_s, c=c, marker=m)
plt.legend()
plt.xlabel('Petal Width (cm)')
plt.ylabel('Probability')
plt.grid()
plt.show()
```

```python
# let's insert another feature into X
X = np.copy(iris['data'][:, 2:])
```

### Decision Boundary: $\theta_{0} + \theta_{1}x_{1} + \theta_{2}x_{2} = 0$

```python
log_reg = LogisticRegression()
log_reg.fit(X, y)

color, marker = [], []
for i in range(len(y)):
    if (y[i] == 0):
        color.append('g')
        marker.append('o')
    else:
        color.append('b')
        marker.append('s')

# representing Petal Length (cm) ==> x1:
x_scatter = np.linspace(4, 6, 100)
# representing Petal Width (cm) ==> x2:
y_scatter = (-1 / log_reg.coef_[0, 1]) * (log_reg.coef_[0, 0] * x_scatter + log_reg.intercept_[0])

plt.figure(figsize=(10.5, 4.5))
plt.plot(x_scatter, y_scatter, c='r', ls='--', label='Decision Boundary')
for x_s, y_s, c, m in zip(X[:, 0], X[:, 1], color, marker):
    plt.scatter(x_s, y_s, c=c, marker=m, s=15)
plt.legend(loc='upper left')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

## Gradient Descent
*virginica* predictor - based on *all 4 features*

### error function: $\dfrac{-1}{m} (y^{T} \ln(\hat{p}) + (A - y)^{T} \ln(A - \hat{p})))$
where $\hat{p} = \dfrac{1}{1 + e^{-X\theta}}$ and $A$ is a $m$x$1$ matrix with all elements equal to zero

---
### gradient of error function: $\dfrac{1}{m} X^{T} (\sigma(X\theta) - y)$
in this case, $\sigma$ represents **sigmoid** function

```python
# data preperation
X = np.copy(iris['data'][:, :])
y = y.reshape(-1, 1)

X_b = np.concatenate((np.ones((150, 1)), X), axis=1)
```

```python
theta = np.random.randn(5, 1)   # random initialization

alpha = 0.01
eta = 0.1   # learning rate
n_iterations = 75000
m = 150

def sigmoid(t):
    return 1 / (1 + np.e ** (-t))

for iteration in range(n_iterations):
    w = np.copy(theta)
    w[0, 0] = 0
    grad = (1 / m) * X_b.T.dot(sigmoid(X_b.dot(theta)) - y) + alpha * w
    theta -= eta * grad
    
print(theta)
```

# Softmax Regression

## Scikit-Learn
*full* predictor - based on *petal length* and *petal width*

```python
X = np.copy(iris['data'][:, 2:])
y = iris.target
```

```python
softmax_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', C=10)
softmax_reg.fit(X, y)
```

```python
xx, yy = np.meshgrid(np.linspace(0, 7, 700), np.linspace(0, 3, 300))
z = softmax_reg.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)

color, marker = [], []
for i in range(len(y)):
    if (y[i] == 0):
        color.append('g')
        marker.append('o')
    elif (y[i] == 1):
        color.append('b')
        marker.append('s')
    else:
        color.append('r')
        marker.append('^')

plt.figure(figsize=(10.5, 4.5))
plt.contourf(xx, yy, z, cmap=plt.cm.Set2)
for x_s, y_s, c, m in zip(X[:, 0], X[:, 1], color, marker):
    plt.scatter(x_s, y_s, c=c, marker=m, s=15)
# plt.legend(loc='upper left')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

## Gradient Descent
*full* predictor - using *all 4 features*

### error function: $\dfrac{-1}{m} y^{T} \ln(\hat{p})$
where $\hat{p} = \sigma(X\theta)$

---
### gradient of error function: $\dfrac{1}{m} X^{T} (\sigma(X\theta) - y)$
in this case, $\sigma$ represents **softmax** function; $\sigma(s(x))_{k} = \dfrac{s_{k}(x)}{\sum{s(x)}}$

```python
# data preperation
X = np.copy(iris['data'][:, :])
y = iris.target

def y_transform(x):
    result = np.zeros((3, ))
    result[x] = 1
    return result

y = map(y_transform, y)
y = np.array(list(y))

X_b = np.concatenate((np.ones((150, 1)), X), axis=1)
```

```python
theta = np.random.randn(5, 3)   # random initialization

alpha = 0
eta = 1e-5   # learning rate
n_iterations = 40000
m = 150
k = 3

def softmax(T):
    result = np.copy(T)
    for i in range(m):
        sum_row = T[i, :].sum()
        for j in range(k):
            result[i, j] = T[i, j] / sum_row
    return result

for iteration in range(n_iterations):
    w = np.copy(theta)
    w[0, 0] = 0
    grad = X_b.T.dot(softmax(X_b.dot(theta)) - y) + alpha * w
    theta -= eta * grad

print(theta)
```

```python
y_prob = softmax(X_b.dot(theta))

def prob_transform(P):
    pred = np.zeros((P.shape[0], 1))
    for i in range(P.shape[0]):
        pred[i, 0] = P[i, :].argmax()
    return pred

y_pred = prob_transform(y_prob)
print(sum(y_pred == iris.target.reshape(-1, 1))[0] / m * 100, "%")
```
