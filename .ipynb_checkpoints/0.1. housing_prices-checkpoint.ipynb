```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
```

## reading data

```python
data = pd.read_table('CaliforniaHousing/cal_housing.data', sep=',')
data
```

```python
data.info()
```

```python
data.describe()
```

```python
data.hist(bins=50, figsize=(20, 15))
plt.show()
```

## Train-Test split

### purely random

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)
```

```python
data['income_category'] = pd.cut(data['medianIncome'], bins=[0., 1.5, 3., 4.5, 6., np.inf],
                                labels=[1, 2, 3, 4, 5])
data['income_category'].hist()
plt.show()
```

### stratified split

```python
from sklearn.model_selection import StratifiedShuffleSplit
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data, data['income_category']):
    strat_train_set = data.loc[train_index]
    strat_test_set = data.loc[test_index]
strat_train_set
```

```python
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_category', axis=1, inplace=True)
strat_train_set
```

```python
data = strat_train_set.copy()
data
```

## Visualizing data

```python
data.plot(kind="scatter", x="longitude", y="latitude", alpha=0.4,
s=data["population"]/100, label="population", figsize=(10,7),
c="medianHouseValue", cmap=plt.get_cmap("jet"), colorbar=True)
plt.legend()
plt.show()
```

```python
data["rooms_per_household"] = data["totalRooms"] / data["households"]
data["bedrooms_per_room"] = data["totalBedrooms"] / data["totalRooms"]
data["population_per_household"] = data["population"] / data["households"]
corr_matrix = data.corr()
plt.subplots(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True)
```

## First Model - Linear Regression

```python
X_train = data.drop('medianHouseValue', axis=1)
y_train = data['medianHouseValue']
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
some_data = X_train.iloc[:5]
some_labels = y_train.iloc[:5]
print("Predictions: ", lin_model.predict(some_data))
print("Labels: ", some_labels)
```

```python
from sklearn.metrics import mean_squared_error
predict_train = lin_model.predict(X_train)
mse_train = mean_squared_error(y_train, predict_train)
rmse_train = np.sqrt(mse_train)
print("Train set RMSE: ", rmse_train) # underfitted
```

## Second Model - Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
tree_model = DecisionTreeRegressor()
tree_model.fit(X_train, y_train)
```

```python
print("Predictions: ", tree_model.predict(some_data))
print("Labels: ", some_labels)
```

```python
predict_train = tree_model.predict(X_train)
mse_train = mean_squared_error(y_train, predict_train)
rmse_train = np.sqrt(mse_train)
print("Train set RMSE: ", rmse_train) # overfitted
```

## Cross-Validation - Decision Tree Regression

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(tree_model, X_train, y_train, scoring="neg_mean_squared_error", cv=10)
rmse_validation = np.sqrt(-scores)

print("Scores: ", scores)
print("Mean: ", scores.mean())
print("Standard Deviation: ", scores.std())
```

## Cross-Validation - Linear Regression

```python
scores = cross_val_score(lin_model, X_train, y_train, scoring="neg_mean_squared_error", cv=10)
rmse_validation = np.sqrt(-scores)

print("Scores: ", scores)
print("Mean: ", scores.mean())
print("Standard Deviation: ", scores.std())
```

## Third Model - Random Forest Regression (Cross-Validation)

```python
from sklearn.ensemble import RandomForestRegressor
forest_model = RandomForestRegressor()
forest_model.fit(X_train, y_train)
```

```python
predict_train = forest_model.predict(X_train)
mse_train = mean_squared_error(y_train, predict_train)
rmse_train = np.sqrt(mse_train)
print("Train set RMSE: ", rmse_train)
```

```python
scores = cross_val_score(forest_model, X_train, y_train, scoring="neg_mean_squared_error", cv=10)
rmse_validation = np.sqrt(-scores)

print("Scores: ", scores)
print("Mean: ", scores.mean())
print("Standard Deviation: ", scores.std())
```
