# Linear Regression - implementation

## data manipulation

```python
import numpy as np, matplotlib.pyplot as plt
```

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

plt.scatter(X, y, c='b', s=15)
plt.show()
```

## Normal Equation

### equation: $\hat{\theta} = (X^{T}X)^{-1} X^{T} y$
### prediction: $\hat{y} = X\theta$

```python
X_b = np.concatenate((np.ones((100, 1)), X), axis=1)
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta
```

```python
X_new = np.array([[0], [2]])
X_new_b = np.concatenate((np.ones((2, 1)), X_new), axis=1)
y_predict = X_new_b.dot(theta)
print(y_predict)
```

```python
plt.scatter(X, y, c='b', s=15)
plt.plot(X_new, y_predict, 'r-')
plt.show()
```

```python
# comparing with scikit-learn
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
print(lin_reg.intercept_, lin_reg.coef_)
```

## Batch Gradient Descent

### error function: $MSE(\theta) = \dfrac{1}{m}(X\theta - y)^{T}(X\theta - y)$
### gradient of error function: $\triangledown_\theta = \dfrac{2}{m}X^{T}(X\theta - y)$

```python
theta = np.random.randn(2, 1)   # random initialization

eta = 0.1   # learning rate
n_iterations = 1000
m = 100

for iteration in range(n_iterations):
    grad = (2 / m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= eta * grad

print(theta)
```

## Stochastic Gradient Descent

```python
theta = np.random.randn(2, 1)   # random initialization

n_epochs = 50
t0, t1 = 5, 50   # learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

for epoch in range(n_epochs):
    for i in range(m):
        random_index = np.random.randint(m)
        xi = X_b[random_index:random_index + 1]
        yi = y[random_index:random_index + 1]
        grad = 2 * xi.T.dot(xi.dot(theta) - yi)
        eta = learning_schedule(m * epoch + i)
        theta -= eta * grad
        
print(theta)
```
